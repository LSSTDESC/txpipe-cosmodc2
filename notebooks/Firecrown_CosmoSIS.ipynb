```python
import sacc
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import astropy.io.fits as pf
```

```python
Ccl_dir = '/global/homes/c/chihway/TXPipe/'
Cosmosis_dir = '/global/homes/c/chihway/txpipe-cosmodc2/cosmosis_config/test_output/'
```

```python
for i in range(4):
    for j in range(i,4):
        ell_cosmosis = np.loadtxt(Cosmosis_dir+'shear_cl/ell.txt')
        cl_cosmosis = np.loadtxt(Cosmosis_dir+'shear_cl/bin_'+str(j+1)+'_'+str(i+1)+'.txt')
        ell_ccl = np.load(Ccl_dir+"cls('source_"+str(i)+"', 'source_"+str(j)+"').npz")['ell']
        cl_ccl = np.load(Ccl_dir+"cls('source_"+str(i)+"', 'source_"+str(j)+"').npz")['cl']
        plt.figure()
        plt.loglog(ell_cosmosis, ell_cosmosis*cl_cosmosis, label='cosmosis')
        plt.loglog(ell_ccl, ell_ccl*cl_ccl, label='ccl')
        plt.title(str(i)+str(j))
        plt.xlabel('ell')
        plt.ylabel('ell*Cell_shear')
        plt.ylim(1e-8, 1e-6)
        plt.xlim(1,10000)
        plt.legend()
        
```

```python
for i in range(4):
    ell_cosmosis = np.loadtxt(Cosmosis_dir+'galaxy_shear_cl/ell.txt')
    cl_cosmosis = np.loadtxt(Cosmosis_dir+'galaxy_shear_cl/bin_1_'+str(i+1)+'.txt')
    ell_ccl = np.load(Ccl_dir+"cls('lens_0', 'source_"+str(j)+"').npz")['ell']
    cl_ccl = np.load(Ccl_dir+"cls('lens_0', 'source_"+str(j)+"').npz")['cl']
    plt.figure()
    plt.loglog(ell_cosmosis, ell_cosmosis*cl_cosmosis, label='cosmosis')
    plt.loglog(ell_ccl, ell_ccl*cl_ccl, label='ccl')
    plt.title(str(i)+str(j))
    plt.xlabel('ell')
    plt.ylabel('ell*Cell_galaxy_shear')
#     plt.ylim(1e-8, 1e-6)
    plt.xlim(1,10000)
    plt.legend()
```
