```python
import sacc
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import astropy.io.fits as pf
```

```python
Ccl_dir = '/global/homes/c/chihway/TXPipe/'
Cosmosis_dir = '/global/homes/c/chihway/txpipe-cosmodc2/cosmosis_config/TEST/test_${DATAFILE}_TEST/'
```

```python
for i in range(4):
    for j in range(i,4):
        ell_cosmosis = np.loadtxt(Cosmosis_dir+'shear_cl/ell.txt')
        cl_cosmosis = np.loadtxt(Cosmosis_dir+'shear_cl/bin_'+str(j+1)+'_'+str(i+1)+'.txt')
        ell_ccl = np.load(Ccl_dir+"cls('source_"+str(i)+"', 'source_"+str(j)+"').npz")['ell']
        cl_ccl = np.load(Ccl_dir+"cls('source_"+str(i)+"', 'source_"+str(j)+"').npz")['cl']
        plt.figure()
        plt.loglog(ell_cosmosis, ell_cosmosis*cl_cosmosis)
        plt.loglog(ell_ccl, ell_ccl*cl_ccl)
        plt.title(str(i)+str(j))
        plt.xlabel('ell')
        plt.ylabel('ell*Cell_shear')
        
```

```python
for i in range(4):

    z_cosmosis = np.loadtxt(Cosmosis_dir+'nz_source/z.txt')
    nz_cosmosis = np.loadtxt(Cosmosis_dir+'nz_source/bin_'+str(i+1)+'.txt')
    z_ccl = np.load(Ccl_dir+'nofz_source_'+str(i)+'.npz')['z']
    nz_ccl = np.load(Ccl_dir+'nofz_source_'+str(i)+'.npz')['nz']
    plt.figure()
    plt.plot(z_ccl, nz_ccl)
    plt.plot(z_cosmosis, nz_cosmosis)

    plt.title(str(i))
```
